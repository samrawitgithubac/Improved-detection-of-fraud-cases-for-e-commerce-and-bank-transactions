Task 1 EDA for the credit card dataset: target imbalance, Amount/Time distributions, and basic sanity checks.


```python
from pathlib import Path
import sys

# Ensure repo root is on PYTHONPATH so `import src...` works in Jupyter
sys.path.insert(0, str(Path("..").resolve()))

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from src.data.creditcard_features import clean_creditcard_df

RAW_DIR = Path("../data/raw")
cc_path = RAW_DIR / "creditcard.csv"

cc = pd.read_csv(cc_path)
cc = clean_creditcard_df(cc)

cc.head()
```

```python
# Class imbalance
class_counts = cc["Class"].value_counts().rename_axis("Class").reset_index(name="count")
class_counts["rate"] = class_counts["count"] / class_counts["count"].sum()
class_counts
```

```python
# Amount distribution by class
plt.figure(figsize=(7, 4))
sns.histplot(
    data=cc,
    x="Amount",
    hue="Class",
    bins=60,
    element="step",
    stat="density",
    common_norm=False,
)
plt.title("Amount by Class")
plt.tight_layout()
plt.show()
```
